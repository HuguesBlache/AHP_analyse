```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import plotly.express as px
```

```python
##### Plot expertise

expertise_columns=result.columns[1]
fig = go.Figure()
colors = px.colors.qualitative.D3
fig.add_trace(go.Pie(labels=list(result[expertise_columns].value_counts().index),values=list(result[expertise_columns].value_counts())))
fig.update_traces( textfont_size=20,
                  marker=dict(colors=colors, line=dict(color='#000000', width=1)))

fig.update_layout(legend=dict( orientation="h", yanchor="middle", y=-0.20, x=0.19, title_font_family="Times New Roman",
 ))
fig.update_layout( width=600, height=600)


fig.show()
fig.write_image("images/expert_count.pdf")
```

```python
##### Plot expertise

import plotly.graph_objects as go

expertise_columns=result.columns[2]

pivot_table=result.pivot_table(index=result.columns[1], columns=result.columns[2], aggfunc='size', fill_value=0)
desired_order = ['Intermediate Level', 'Expert Level', 'Advanced Level', 'Beginner Level', 'No expertise']

# Réorganiser les index selon l'ordre désiré
pivot_table = pivot_table.reindex(desired_order)

table_by_expert=[]
for i in range(0,len(pivot_table)):
    table_by_expert.append(list(pivot_table.iloc[i]))

```

```python
fig=go.Figure()
for i in range(0,len(table_by_expert)):
    fig.add_trace(go.Bar(x=pivot_table.columns,y=table_by_expert[i],name=desired_order[i],marker_color=colors[i]))
fig.update_layout(barmode='stack')
fig.update_layout(legend=dict( orientation="h", yanchor="middle", y=-0.20, xanchor="center", x=0.5, title_font_family="Times New Roman",
 ))
fig.update_xaxes(title_text="<b>Scenario</b>")
fig.update_xaxes(title_text="<b>Count</b>")
fig.update_traces( textfont_size=30,
                  marker=dict(line=dict(color='#000000', width=1)))
fig.update_layout( width=800, height=800)
fig.update_layout( plot_bgcolor='rgb(245, 245, 245)')
fig.write_image("images/expert_scenario.pdf")
fig.show()
```

### Test d'un plot

```python
list_comparaison_str=['Extremely in highway','Strongly in highway','A bit in highway','Equality'
                      ,'A bit in highway','Strongly in highway','Extremely in highway']
list_comparaison_num=[1/4,1/2,1/2,1,2,3,4]
list_comparaison_num=1/np.array(list_comparaison_num)
list_answer=list(result[result.columns[-5]])
list_answer_num=[]
for i in list_answer:
    list_answer_num.append(list_comparaison_num[list_comparaison_str.index(i)])
```

```python

q1 = np.percentile(list_answer_num, 18.75)
median = np.median(list_answer_num)
q3 = np.percentile(list_answer_num, 100-18.75)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default='notebook'
```

```python
fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Scatter(x=[q1,median],y=[0,1],mode='lines',line=dict(color='#DC3912')),row=1, col=1)
fig.add_trace(go.Scatter(x=[median,q3],y=[1,0],mode='lines',line=dict(color='#DC3912')),row=1, col=1)


fig.add_trace(go.Box(x=list_answer_num,name="Reponse",marker=dict(
        color='#DC3912'  # Choisir la couleur souhaitée
    )),row=2, col=1)

fig.update_yaxes( 
            showgrid = True,
            
              showline = True,
              gridcolor = 'rgb(200,200,200)',
              gridwidth = 2,
              zerolinecolor = '#969696',
              zerolinewidth = 2,
              linecolor = 'rgb(253, 253, 253)',
              linewidth = 2,
              color ='rgb(55,126,184)',
                        title_text='Fuzzy Number',row=1, col=1
        )
fig.update_xaxes(   showgrid = True,
            
              showline = True,
              gridcolor = 'rgb(200,200,200)',
              gridwidth = 2,
              zerolinecolor = '#969696',
              zerolinewidth = 2,
              linecolor = 'rgb(253, 253, 253)',
              linewidth = 2,
              color ='rgb(55,126,184)',
     title_text='Value',
range=[1-0.2, 4+0.2]
) 

fig.update_layout(
plot_bgcolor= 'rgb(253, 253, 253)',
 showlegend=False ,
title_x=0.5
)

fig.show()
```
